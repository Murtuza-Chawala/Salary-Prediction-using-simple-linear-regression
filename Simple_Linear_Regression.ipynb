### Pre processing the data

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
#Storing the independent variables
X = dataset.iloc[:, :1].values #:  (:1 to convert independent variable in a 2d array as required by the ML model.)
#Storing the dependent variables
y = dataset.iloc[:, 1].values

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

```

```python
print('Independent Variable - Experience(in Yrs)')
print(X)
print('Dependent Variable - Salary')
print(y)
```

###  Training our model, ie Fitting Simple Linear Regression to the training set (*The library we will be using takes care of feature scaling)



```python
#Loading and training our model to find co-relations
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

### Predicting the salaries earned by employees based on on our test_data set (ie X_test)

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

### y_pred is a vector consisting of salaries which are predicted by our ML model which we trained through our training data_set(X_train and y_train).

## Comparing accuracy of our ML model 
### We now have two vectors  1) y_test :- Consisting of salaries earned in real life.  2)y_pred :-Consisting of salaries predicted by our model, for the same set of Input (Experience in yrs). 
### The similarity between y_test(actual) and y_pred is directly proportional to ML model accuracy

```python
#Visualizing the training set


plt.scatter(X_train,y_train, color='red') #Creating a scatter plot
plt.plot(X_train,regressor.predict(X_train), color='green') #Plotting a regression line
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing the test set


plt.scatter(X_test,y_test, color='red') #Creating a scatter plot
plt.plot(X_train,regressor.predict(X_train), color='green') #Plotting a regression line
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

### Predicting a salary (data = hardcoded)

```python
print('The employee who has worked for {} no of years will get an approx salary of {} $'.format(7,round(float(regressor.predict(np.array([[7]]))),3)))
```

### Predicting a salary based on an input year

```python
n = float(input('Please enter employee experience (in no.of years) '))
y = np.array([[n]])
y1_pred = regressor.predict(y)
print()
print('The employee who has worked for {} no of years will get an approx salary of {} $'.format(n,round(float(y1_pred),3)))
```
